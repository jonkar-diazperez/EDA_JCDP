# MEMORIA EDA JCDP - ZBE

## INTRODUCCION

El objetivo de este EDA es analizar si la introducción de las ZBE en las capitales de Euskadi ha tenido un impacto positivo en los indices de calidad del aire (ICA) de esas zonas y las que se encuentran en las proximidades.
Las hipótesis planteadas basadas en este objetivo han sido:
* H1: ¿Se han reducido los ICA en las ZBE?
* H2: ¿Se han reducido todos los ICA o hay ICA con más impacto?
* H2: ¿Han aumentado/disminuido los ICA de las zonas externas a las ZBE? (Efecto Frontera)


## FUENTES DE DATOS
Para hacer este análisis, vamos a usar los CSV con la información de los Indices de Calidad del Aire (ICA) de las estaciones urbanas disponibles para las 3 capitales, para comparar los valores previos a la implantación de las ZBE en cada una de ellas (2024) y los posteriores 2025, que se encuentran disponibles en el sistema de calidad del aire del País Vasco [Air Quality System of the Basque Country](https://opendata.euskadi.eus/catalogo-datos/?r01kQry=tC:euskadi;tT:ds_informes_estudios;m:documentLanguage.EQ.es;mO:documentName.LIKE.calidad%20del%20aire,documentDescription.LIKE.calidad%20del%20aire,OpendataLabels.LIKE.calidad%20del%20aire;cA:r01e00000fe4e66771ba470b85e6897e3cbce045d;pp:r01PageSize.10;p:Inter_portal,Inter&r01SearchEngine=meta).

Esta información está disponible en distintos formatos (XLS, CSV, JSON y API REST), y vienen agrupadas en datos diarios horarios y datos de índices de calidad del aire para cada estación de la red. Para el EDA hemos seleccionado los datasets de las estaciones que estan situadas en las ZBE y en sus proximidades para comparar los valores de los ICA antes y después de implantarlas.

## LIMPIEZAS
Como indicaba anteriormente, el primer paso de este proceso ha sido seleccionar las estaciones que vamos a usar para el EDA y cargar los CSV de esas estaciones de los centros urbanos de Bilbao, Donostia y Gasteiz de los años 2024 y 2025 (excepto en el caso de Gasteiz, ya que la ZBE comenzó a funcionar en Septiembre de 2025).
Tras concatenar estos ficheros, que tienen las mismas columnas de datos para cada estación, eliminamos las filas con valores nulos (por tener registros horarios y la baja ocurrencia de nulos, considero que se pueden despreciar y que no influyen en gran medida en el resultado).

La [Ley 7/2021 de cambio climático y transición energética](https://www.boe.es/diario_boe/txt.php?id=BOE-A-2021-8447), de carácter europeo, obliga a los municipios de más de 50.000 habitantes a que adopten medidas para reducir las emisiones derivadas de la movilidad.

Para compartir las experiencias de la implantación en los ayuntamientos de EUSKADI, se ha constituido un [grupo de trabajo de Udalsarea 2030](https://www.udalsarea2030.eus/noticias/municipios-vascos-comparten-proceso-implantacion-zonas-bajas-emisiones-zbe) en donde tambien participan otras entidades como Ihobe, Eudel y el Departamento de Movilidad Sostenible del Gobierno Vasco.

##  Impacto medioambiental.
La implantación de las ZBE pretende generar una serie de beneficios para los ciudadanos de los municipios:

* *Reducción de la contaminación atmosférica y por lo tanto mejora de la calidad del aire que respiramos.*

* *Reducción de la contaminación acústica*

* *Favorecer e impulsar el transporte público*

* *Promover hábitos más saludables*

##  HIPÓTESIS.
El objetivo de este trabajo es evaluar el impacto que ha tenido hasta ahora la implantación de estas zonas en las capitales vascas desde el punto de vista medioambiental, usando los datos de las mediciones de calidad del aire que se realizan diariamente en las estaciones del sistema [Air Quality System of the Basque Country](https://opendata.euskadi.eus/catalogo-datos/?r01kQry=tC:euskadi;tT:ds_informes_estudios;m:documentLanguage.EQ.es;mO:documentName.LIKE.calidad%20del%20aire,documentDescription.LIKE.calidad%20del%20aire,OpendataLabels.LIKE.calidad%20del%20aire;cA:r01e00000fe4e66771ba470b85e6897e3cbce045d;pp:r01PageSize.10;p:Inter_portal,Inter&r01SearchEngine=meta).

* H1: ¿Se han reducido los ICA en las ZBE?
* H2: ¿Se han reducido todos los ICA o hay ICA con más impacto?
* H3: ¿Han aumentado/disminuido los ICA de las zonas externas a las ZBE? (Efecto Frontera)


Para hacer esta evaluación, vamos a usar los CSV con la información de los Indices de Calidad del Aire (ICA) de las estaciones urbanas disponibles para las 3 capitales, para comparar los valores previos a la implantación de las ZBE en cada una de ellas (2024) y los posteriores 2025:
